```python
def median_calculator(df, col):
    return df[col].median()
median_bmi = median_calculator(df_runners, "BMI")
v02_median = median_calculator(df_runners, "VO2 Max")
cadence_median  = median_calculator(df_runners, "Cadence")
median_age = median_calculator(df_runners, "Age")

df_runners["BMI"] = df_runners["BMI"].fillna(median_bmi)
df_runners["VO2 Max"] = df_runners["VO2 Max"].fillna(v02_median)
df_runners["n Marathons run"]= df_runners["n Marathons run"].fillna(0)
df_runners["Cadence"] = df_runners["Cadence"].fillna(cadence_median)
df_runners["Age"] = df_runners["Age"].fillna(median_age)
```
